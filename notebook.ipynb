### Import Project Utilities

## 2. Main Code

### Initialize Model

```python
# Create model
model = SimpleNN(input_size=10, hidden_size=64, output_size=2)
model = model.to(device)

print(f"✓ Model created and moved to {device}")
print(f"Model parameters: {sum(p.numel() for p in model.parameters()):,}")
print(f"\n{model}")
```

### Create Sample Data

```python
# Create sample data
from torch.utils.data import TensorDataset, DataLoader

# Generate random data for demonstration
X_train = torch.randn(1000, 10)
y_train = torch.randint(0, 2, (1000,))

train_dataset = TensorDataset(X_train, y_train)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

print(f"✓ Dataset created with {len(train_dataset)} samples")
print(f"✓ Batch size: 32")
```

### Train Model

```python
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm

# Training setup
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
num_epochs = 5

# Training loop
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    
    for batch_x, batch_y in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}"):
        batch_x, batch_y = batch_x.to(device), batch_y.to(device)
        
        optimizer.zero_grad()
        outputs = model(batch_x)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
    
    avg_loss = total_loss / len(train_loader)
    print(f"Epoch {epoch+1}/{num_epochs} - Loss: {avg_loss:.4f}")

print("✓ Training completed!")
```

### Save Model

```python
# Save model
model_path = "model.pth"
torch.save(model.state_dict(), model_path)
print(f"✓ Model saved to {model_path}")

# If in Colab, save to Google Drive
if in_colab:
    import shutil
    shutil.copy(model_path, "/content/drive/MyDrive/model.pth")
    print(f"✓ Model also saved to Google Drive")
```
